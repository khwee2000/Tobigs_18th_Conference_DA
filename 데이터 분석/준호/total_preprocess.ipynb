## 2017년 

```python
import pandas as pd
s=pd.read_csv('seoul_moving_data/2017/LOCAL_PEOPLE_GU_201701.csv',encoding='cp949')
code_area=s['자치구코드'].unique()

months=['01','02','03','04','05','06','07','08','09','10','11','12']
# create empty dataframe with columns same as 's'

temp_2017=pd.DataFrame(columns=s.columns)


    
data_2017=[]
for code in code_area[1:]:
    temp=pd.DataFrame(columns=s.columns)
    for month in months:
        data=pd.read_csv('seoul_moving_data/2017/LOCAL_PEOPLE_GU_2017'+month+'.csv',encoding='cp949')
        data=data[data['자치구코드']==code]
        temp=pd.concat([temp,data],axis=0)

    ## save in csv
    data_2017.append(temp)
```

## 2018 년


```python
import pandas as pd
import numpy as np
months=['01','02','03','04','05','06','07','08','09','10','11','12']

def read_per_month(month):
    data=pd.read_csv('seoul_moving_data/2018/LOCAL_PEOPLE_GU_2018'+'01'+'.csv',encoding='cp949')
    data.reset_index(drop=False,inplace=True)
    x=data.columns[1:]
    x=x.tolist()
    x.append("temp")
    data.columns=x
    data.drop('temp',axis=1,inplace=True)
    return data
new_data_per_month=[]
for month in months:
    data=read_per_month(month)
    new_data_per_month.append(data)
data_2018=[]
for code in code_area[1:]:
    temp=pd.DataFrame(columns=data.columns)
    for month in months:
        data=new_data_per_month[int(month)-1]
        data=data[data['자치구코드']==code]
        temp=pd.concat([temp,data],axis=0)
    data_2018.append(temp)
```
