```python
import pandas as pd
import numpy as np
months=['01','02','03','04','05','06','07','08','09','10','11','12']


def read_per_month(month):
    data=pd.read_csv('seoul_moving_data/2018/LOCAL_PEOPLE_GU_2018'+'01'+'.csv',encoding='cp949')
    data.reset_index(drop=False,inplace=True)
    x=data.columns[1:]
    x=x.tolist()
    x.append("temp")
    data.columns=x
    data.drop('temp',axis=1,inplace=True)
    return data
```

```python
new_data_per_month=[]
for month in months:
    data=read_per_month(month)
    new_data_per_month.append(data)
    
```

```python
code_area=data['자치구코드'].unique()
results=[]
for code in code_area[1:]:
    temp=pd.DataFrame(columns=data.columns)
    for month in months:
        data=new_data_per_month[int(month)-1]
        data=data[data['자치구코드']==code]
        temp=pd.concat([temp,data],axis=0)
    results.append(temp)
```
